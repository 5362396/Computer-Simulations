Napisać funkcję obliczającą kolejne iteracje dla dyskretnego modelu wzrostu logistycznego. Parametry r, x0 oraz n (ilość iteracji) ustalić jako argumenty funkcji. Korzystając z tej funkcji:

```python
import matplotlib.pyplot as plt
import random


def wzrost_logistyczny(r, x, n):
    x_list = [x]
    for _ in range(n):
        x = r * x * (1 - x)
        x_list.append(x)
    return x_list
```

Zad. 1. Zachowanie iteracji logistycznej:

(a) Wygenerować wykresy ciągu (xt) dla t = 0,...,100, r = 1.1 oraz pięciu losowych wartości x0 ∈ (0,1).

```python
x = [i for i in range(101)]
```

```python
for _ in range(5):
    rnd = random.uniform(0, 1)
    y = wzrost_logistyczny(1.1, rnd, 100)
    plt.plot(x, y)
```

Zad. 2. Mapa punktów:

(a) Wygenerować mapę punktów dla dyskretnego modelu logistycznego dla ustalonego x0 oraz r ∈ [2.8,4] (zgodnie z opisanym na wykładzie algorytmem).

```python
r = 2.8
xo = 0.5

while r < 4.001:
    y = wzrost_logistyczny(r, xo, 300)
    y = y[250:]
    plt.plot([r for _ in range(51)], y, 'bd', markersize=1)
    r += 0.004
```

(b) Jak brak odrzucenia początkowych wartości xt wpływa na mapę punktów?

```python
r = 2.8
xo = 0.5

while r < 4.001:
    y = wzrost_logistyczny(r, xo, 300)
    plt.plot([r for _ in range(301)], y, 'bd', markersize=1)  # wartości początkowe są niestabilne
    r += 0.004
```

Zad. 3. Wpływ wartości początkowych na zachowanie układu:

(a) Jak wybór x0 wpływa na zmianę mapy punktów?

```python
r = 2.8
xo = 0.99  # nie wpływa

while r < 4.001:
    y = wzrost_logistyczny(r, xo, 300)
    y = y[250:]
    plt.plot([r for _ in range(51)], y, 'bd', markersize=1)
    r += 0.004
```

Zad. 4. Okresowe cykle:

(a) Wyznaczyć wartości r, dla których pojawiają się okresowe cykle (np. cykl o długości 2, 4, 8).

```python
r = 2.8
xo = 0.5
cykle = []
ind = 1

while r < 4.001:
    y = wzrost_logistyczny(r, xo, 300)
    y = y[250:]
    y_rounded = [round(y, 2) for y in y]
    if len(set(y_rounded)) > ind:
        cykle.append(r)
        ind *= 2
    plt.plot([r for _ in range(51)], y, 'bd', markersize=1)
    r += 0.004

ind = 1
for r in cykle:
    ind *= 2
    print(f'Cykl o długości {ind}: r = {r}')
```
