```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def regula_90(initial_state, number_of_generations):
    number_of_cells = len(initial_state)
    grid = np.zeros((number_of_generations, number_of_cells), dtype=int)
    grid[0] = initial_state

    for generation in range(1, number_of_generations):
        for i in range(1, number_of_cells - 1):
            grid[generation, i] = grid[generation-1, i-1] ^ grid[generation-1, i+1]
        grid[generation, 0] = grid[generation-1, 1]
        grid[generation, -1] = grid[generation-1, -2]
    return grid
```

```python
def rysuj_generacje(grid):
    plt.figure(figsize=(10, 10))
    plt.imshow(grid, cmap='binary', interpolation='nearest')
    plt.title('Automat komórkowy Reguła 90')
    plt.xlabel('Index komórki')
    plt.ylabel('Generacja')
    plt.show()
```

```python
num_cells = 95
num_generations = 48
initial_state = [0] * num_cells
initial_state[num_cells // 2] = 1

grid = regula_90(initial_state, num_generations)
rysuj_generacje(grid)
```
